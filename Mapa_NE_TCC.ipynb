```python
# Instalação e importação das bibliotecas
!pip install geopandas matplotlib
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.lines import Line2D
from google.colab import files

# Carregamento do shapefile e configuração da plotagem
uploaded = files.upload()
uf_ibge = 'estados.shp'
brasil = gpd.read_file(uf_ibge)
nordeste_estados = ['MA', 'PI', 'CE', 'RN', 'PB', 'PE', 'AL', 'SE', 'BA']
nordeste = brasil[brasil['sigla'].isin(nordeste_estados)]

# Configuração e plotagem gráfico
fig, ax = plt.subplots(figsize=(12, 12))
brasil.boundary.plot(ax=ax, linewidth=1, color='black')
nordeste.plot(ax=ax, color='#FFCC80')
ax.set_title('Região Nordeste do Brasil', fontsize=15)
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
brasil_patch = Line2D([0], [0], color='black', linewidth=1, label='Brasil')
nordeste_patch = mpatches.Patch(color='#FFCC80', label='Nordeste')
ax.legend(handles=[brasil_patch, nordeste_patch], loc='upper right')

# Definição do formato e exibição
plt.savefig('nordeste_brasil.jpg', format='jpg', dpi=400)
plt.show()
```
